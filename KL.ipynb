```python
import numpy as np
from math import log2
from PIL import Image
```

```python
import tensorflow as tf
cifar = tf.keras.datasets.cifar10
(x_train, _), (x_test, _) = cifar.load_data()
```

```python
image1 = x_train[5]
image2 = x_train[5]

array1 = np.array(image1)
array2 = np.array(image2)
```

```python
# histograms and normalize to obtain probability distributions
hist1, _ = np.histogram(array1, bins=256, range=(0, 256), density=True)
hist2, _ = np.histogram(array2, bins=256, range=(0, 256), density=True)

hist1 = hist1[:256]
hist2 = hist2[:256]
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_range = np.arange(256)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0, 0].plot(x_range, hist1, color='blue')
axes[0, 0].set_title('Density of hist1')
axes[0, 0].set_xlabel('Pixel Value')
axes[0, 0].set_ylabel('Density')

axes[0, 1].hist(np.arange(256), bins=256, weights=hist1, color='blue')
axes[0, 1].set_title('Histogram of hist1')
axes[0, 1].set_xlabel('Pixel Value')
axes[0, 1].set_ylabel('Frequency')

axes[1, 0].plot(x_range, hist2, color='red')
axes[1, 0].set_title('Density of hist2')
axes[1, 0].set_xlabel('Pixel Value')
axes[1, 0].set_ylabel('Density')

axes[1, 1].hist(np.arange(256), bins=256, weights=hist2, color='red')
axes[1, 1].set_title('Histogram of hist2')
axes[1, 1].set_xlabel('Pixel Value')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
def kl_Loss(p, q, epsilon=1e-10):
    pq=sum(p[i] * log2((p[i] + epsilon) / (q[i] + epsilon)) for i in range(len(p)))
    qp=sum(q[i] * log2((q[i] + epsilon) / (p[i] + epsilon)) for i in range(len(q)))
    p='image1'
    q='imag2'
    print(" {},differs from another distribution, {}: {}\n\n".format(p,q,pq))
    print(" {},differs from another distribution, {}: {}\n\n".format(q,p,qp))
    return (pq+qp)

kl_score = kl_Loss(hist1, hist2)
print('KL Loss Score(sum of both score by swapping): %.5f pixel' % kl_score)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(image1)
axes[0].set_title('Image 1')
axes[1].imshow(image2)
axes[1].set_title('Image 2')
plt.show()
```
